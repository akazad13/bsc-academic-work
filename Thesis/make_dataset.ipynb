```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import os
from PIL import Image
from numpy import *
# SKLEARN
from sklearn.utils import shuffle
from sklearn.cross_validation import train_test_split
```

```python
# input image dimensions
img_rows, img_cols = 64, 64

# number of channels
img_channels = 3

#  data

path1 = '/home/mirror/Documents/pic'    #path of folder of images    
path2 = '/home/mirror/Documents/out'  #path of folder to save images

```

 ##### For open image and rotate ####

### for resize image ###

```python
listing = sorted(os.listdir(path1))
num_samples=size(listing)
print (num_samples)


for file in listing:
    im = Image.open(path1 + '/' + file)
    print(im.size)
    img = im.resize((img_rows,img_cols))
    
    print(file)
    
    img.save(path2 +'/' +  file, "JPEG")
    
imlist = sorted(os.listdir(path1))
```

```python
immatrix = array([array(Image.open(path2+ '/' + im2))
              for im2 in imlist],'f')

print(immatrix.size)
```

```python
label=np.ones((num_samples,),dtype = int)
label[0:5]=0
label[5:10]=1

for x in range(0, 10):
    img=immatrix[x].reshape(img_rows,img_cols,img_channels)
    plt.imshow(img)
    print(label[x])
    
```

```python
data,Label = shuffle(immatrix,label, random_state=2)
train_data = [data,Label]
```
